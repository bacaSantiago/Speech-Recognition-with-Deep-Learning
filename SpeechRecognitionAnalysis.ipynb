<p><h4> <b>3. Theoretical Framework</b> </h4></p><br>


**3.1 Speech Recognition Fundamentals**  

Speech recognition, or automatic speech recognition (ASR), is the process by which spoken language is converted into text. ASR systems rely on extracting phonetic and linguistic features from audio data, mapping these features to corresponding words or sentences using statistical models or machine learning (Rabiner & Juang, 1993).

<br>

**3.2 Deep Learning for ASR**  

Deep learning has revolutionized ASR by introducing models such as CNNs and LSTMs, which can capture both local and temporal dependencies in speech. These models have significantly improved the accuracy and efficiency of speech recognition systems (Graves et al., 2013).

- **CNNs**: Convolutional neural networks extract spatial features from spectrograms, a visual representation of sound.
- **LSTMs**: Long short-term memory networks handle sequential data, making them suitable for ASR tasks that require memory of previous inputs (Hochreiter & Schmidhuber, 1997).

<br>

**3.3 Acoustic Features for Speech Recognition**  

One of the most common features extracted from speech is the Mel-frequency cepstral coefficients (MFCCs), which capture the power spectrum of the sound (Davis & Mermelstein, 1980). These features are used in both classical models and deep learning approaches, as they are highly effective in capturing the characteristics of human speech.

<br>

**3.4 Bayesian Networks**  

Bayesian Networks (BNs) are probabilistic graphical models used to represent a set of variables and their conditional dependencies (Pearl, 1988). In the context of speech recognition, Bayesian networks introduce probabilistic reasoning, helping improve decision-making by modeling the relationships between different phonemes, words, and linguistic features. This probabilistic approach enables the ASR system to better handle uncertainty and contextual information, especially in noisy environments (Pearl, 1988).

<br>

**3.5 The Hybrid Approach: DNN + Bayesian Networks**  

Neural networks, particularly LSTMs, have emerged as a powerful tool for ASR due to their ability to model complex temporal relationships in speech data (Hinton et al., 2012). In contrast to traditional ASR systems that rely on acoustic models and phonetic dictionaries, LSTMs learn directly from raw audio features, such as Mel-frequency cepstral coefficients (MFCCs).

The hybrid approach presented in this paper combines DNNs for feature extraction with Bayesian networks for probabilistic reasoning. This method provides both the powerful pattern recognition capabilities of DNNs and the flexibility of Bayesian networks to manage uncertainty (Bengio et al., 2003).

<br>

**3.6 SpeechRecognition Library**  

The SpeechRecognition library for Python offers a simple interface to implement speech recognition. It supports various engines, including Google Web Speech API and CMU Sphinx, a powerful offline recognition engine (Zhang & Wu, 2020). Sphinx operates without an internet connection, which makes it a viable option for offline applications. However, the recognition accuracy of Sphinx is often lower than state-of-the-art deep learning models, particularly in noisy environments.

Given its ease of use and integration capabilities, SpeechRecognition is widely used in lightweight applications where simplicity and fast deployment are prioritized over accuracy. Despite its limitations, the library remains a valuable tool for developers working on speech-to-text applications that do not require high levels of precision.

For the SpeechRecognition approach, pre-recorded audio files were processed using the **CMU Sphinx** recognizer. **4.2. Neural Network Model**

The neural network model was constructed using TensorFlow and Keras. The architecture consists of several layers:

- **Input Layer**: Accepts 13-dimensional MFCC feature vectors as input.
- **LSTM Layer**: A single LSTM layer with 128 units that captures the temporal relationships in the speech data.
- **Dense Layers**: Two fully connected dense layers with 64 and 32 units, respectively, using ReLU activation functions.
- **Output Layer**: A softmax layer that outputs a probability distribution over the possible classes (i.e., words).

In general, the DNN processes audio input to extract features, which are then refined by the Bayesian network using probabilistic reasoning based on linguistic context. The DNN was trained using TensorFlow, and the Bayesian network was constructed using the pgmpy library in Python. The model was compiled using the **Adam optimizer** and **categorical crossentropy loss**.

<p><h4> <b>7. Future Research</b> </h4></p><br>

**7.1 Recommendations for Model Improvements**

The results of this study highlight several areas where future research can focus on improving speech recognition models. Specifically:

- **Noise Robustness in Speech Recognition Models**: While the LSTM neural network performed well in clean environments, further work is required to enhance its ability to handle noisy conditions. Techniques such as **data augmentation** (adding artificial noise to training data) and **spectral noise reduction** (filtering background noise from audio input) could significantly improve robustness in real-world environments. Integrating **attention mechanisms** in LSTM models could also allow the network to focus on more relevant parts of the input sequence and ignore irrelevant noise.
  
- **Hybrid Approaches**: One promising direction for future research is to explore hybrid models that combine deep learning architectures (such as LSTM or CNN) with traditional speech recognition systems. For example, a hybrid model that uses CMU Sphinx for quick initial recognition and LSTM for refinement could capitalize on the strengths of both approaches. This might reduce computational cost while maintaining high accuracy in complex or noisy environments.

- **Incorporation of Transformer-based Architectures**: Transformer models, particularly those used in natural language processing (e.g., BERT, GPT), have demonstrated superior performance in sequence-to-sequence tasks. Applying **transformer-based architectures** to speech recognition, either in isolation or in combination with LSTMs, could lead to significant improvements in both accuracy and robustness, especially for longer sequences of speech input.

<br>

**7.2 Limitations and Disadvantages**

While the LSTM neural network approach has demonstrated superior accuracy, it comes with certain **disadvantages**:

- **Computational Complexity**: Training deep learning models, especially LSTMs, is computationally expensive and requires significant resources. For real-time applications, this limitation could result in slower response times or the need for specialized hardware, such as GPUs.

- **Data Requirements**: Neural networks typically require large amounts of labeled data for training. The performance of the LSTM model could degrade if applied to smaller datasets or languages with fewer available speech corpora. Future work should explore techniques like **transfer learning** or **semi-supervised learning** to mitigate this dependency on large datasets.

In contrast, the SpeechRecognition library, though lightweight and easier to implement, suffers from a major drawback—**poor performance in noisy environments**. While suitable for simple tasks in controlled settings, it cannot be reliably applied to real-world use cases without substantial modifications or improvements.

<br>

**7.3 Future Directions for Research**

Based on the results and limitations observed, future research should focus on the following directions:

- **Integration of Noise Reduction Techniques**: Incorporating noise reduction mechanisms, such as **spectral subtraction**, **Wiener filtering**, or deep learning-based **denoising autoencoders**, could help mitigate the impact of background noise on both models.
  
- **Exploration of Multimodal Systems**: Research could also explore multimodal approaches, combining speech recognition with **visual information** (such as lip reading) or **contextual data** (from the surrounding environment) to improve accuracy in challenging conditions.

- **Evaluation on Multilingual and Dialectal Datasets**: Future research should also expand the evaluation to **multilingual** and **dialectal** speech datasets to determine the generalizability of these models across different languages and accents.

- **Use of Spelling Correction Methods**: Implement spelling correction techniques such as stemming and lemmatization in the SpeechRecognition library model to improve the accuracy of recognized text, especially in challenging noise environments.

**10.1 Neural Network Approach**

This section describes the deep learning approach used to classify audio data using a neural network. The key components include the following:

- **librosa**: A Python library used to process audio files and extract features such as MFCC (Mel-Frequency Cepstral Coefficients). MFCCs are a crucial feature for representing the power spectrum of audio data.
  
- **Neural Network Architecture**: The model consists of an LSTM layer, which is well-suited for sequential data such as audio, followed by a Flatten layer and Dense layers for classification. The input shape corresponds to the MFCC features extracted from the audio files.

- **Audio Processing**: The function `load_data` extracts MFCC features from audio files located in the `LibriSpeech` dataset. It loads each file, converts the audio to a 13-MFCC representation, and assigns the appropriate label using the `LabelBinarizer`.

- **Model Compilation and Training**: After loading and preprocessing the data, the neural network model is compiled using `categorical_crossentropy` as the loss function and the Adam optimizer. The model is then trained on the training data using a batch size of 32 for 10 epochs.

- **Evaluation and Results**: The model is evaluated using precision, recall, and accuracy metrics on the test set, providing insight into the model's ability to classify audio data correctly.

```python
# Import the necessary libraries
import os
import numpy as np
import librosa
import random
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Flatten, Input
from tensorflow.keras.metrics import Precision, Recall
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split

# Set random seeds for reproducibility
random.seed(81)
np.random.seed(81)
tf.random.set_seed(81)

# Function to load the audio data
def load_data(data_dir, sample_rate=16000, max_files_per_class=None):
    X = []
    y = []
    labels = sorted(os.listdir(data_dir))

    # Initialize the LabelBinarizer
    label_binarizer = LabelBinarizer()
    labels = [label for label in labels if label != '.DS_Store']  # Remove `.DS_Store` from labels
    label_binarizer.fit(labels)

    # Iterate through the data directory and load the data
    for label in labels:
        label_dir = os.path.join(data_dir, label)
        if not os.path.isdir(label_dir):
            continue
        
        # Get all the valid audio files
        files = [os.path.join(root, file)
                 for root, _, file_list in os.walk(label_dir)
                 for file in file_list
                 if file.endswith('.wav') or file.endswith('.flac')]

        # Load only the first `max_files_per_class` files
        for file in files[:] if max_files_per_class is None else files[:max_files_per_class]:
            try:
                audio, sr = librosa.load(file, sr=sample_rate)
                mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
                mfccs = np.mean(mfccs.T, axis=0)
                X.append(mfccs)
                y.append(label)
            except Exception as e:
                print(f"Error processing {file}: {e}")

    # Check if any valid data was found
    if len(X) == 0 or len(y) == 0:
        print("No valid audio files found. Exiting...")
        return None, None, None

    # Transform labels using the LabelBinarizer
    y = label_binarizer.transform(y)

    return np.array(X), np.array(y), label_binarizer

# Load the full dataset
X, y, label_binarizer = load_data('assets/LibriSpeech/test-clean/')

# Split the data into training and testing sets, ensuring class balance
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, stratify=y, random_state=81)
print(f"Number of training samples: {len(X_train)}")
print(f"Number of test samples: {len(X_test)}")

if X_train is None or y_train is None:
    print("Exiting due to empty training dataset.")
else:
    # One-hot encode the labels for categorical crossentropy loss
    y_train = np.array([np.array(xi) for xi in y_train])  # Convert list of arrays to 2D array
    y_test = np.array([np.array(xi) for xi in y_test])

    # Define the LSTM model
    model = Sequential([
        Input(shape=(1, 13)),  # Use Input layer to fix the warning
        LSTM(128, return_sequences=True),  # Keep return_sequences=True
        Flatten(),  # Add a Flatten layer to remove the extra dimension
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),
        Dense(len(y_train[0]), activation='softmax')  # Use softmax for multi-class classification
    ])

    # Compile and fit the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy', Precision(), Recall()])
    model.fit(X_train.reshape(X_train.shape[0], 1, X_train.shape[1]), y_train, epochs=10, batch_size=32)

    # Evaluate the model
    loss, accuracy, precision, recall = model.evaluate(X_test.reshape(X_test.shape[0], 1, X_test.shape[1]), y_test)
    print(f'\nTest Accuracy: {accuracy*100:.2f}%', f'\nTest Precision: {precision*100:.2f}%', 
          f'\nTest Recall: {recall*100:.2f}%')
```

**10.2 SpeechRecognition Library Approach**

The second approach utilizes the SpeechRecognition library in Python, which provides access to different speech recognition engines, including the offline Sphinx engine.

- **SpeechRecognition Library**: This Python library handles speech-to-text conversion using various recognition engines like Sphinx (for offline processing). It enables easy access to audio recording and transcription.

- **Word-Level Metrics**: This function calculates the precision, recall, and accuracy of the speech recognition system at the word level. By comparing the expected and recognized phrases word by word, it provides a more granular analysis of the model's performance.

- **Speech Recognition**: The `fn_speech_recognition_wav` function processes a list of `.wav` files, using the Sphinx engine to convert audio into text. It calculates word-level accuracy, precision, and recall based on the comparison between recognized and expected phrases.

```python
# Import the necessary libraries
import speech_recognition as sr
from collections import Counter

# Function to calculate word-level metrics
def calculate_word_level_metrics(expected_phrases, recognized_phrases):
    correct_word_count = 0
    total_expected_words = 0
    total_recognized_words = 0

    for expected, recognized in zip(expected_phrases, recognized_phrases):
        expected_words = expected.lower().split()
        recognized_words = recognized.lower().split()

        # Count correct words
        expected_counter = Counter(expected_words)
        recognized_counter = Counter(recognized_words)
        
        common_words = expected_counter & recognized_counter
        correct_word_count += sum(common_words.values())

        # Total words in expected and recognized phrases
        total_expected_words += len(expected_words)
        total_recognized_words += len(recognized_words)

    # Calculate word-level precision, recall, and accuracy
    precision = correct_word_count / total_recognized_words if total_recognized_words > 0 else 0
    recall = correct_word_count / total_expected_words if total_expected_words > 0 else 0
    accuracy = correct_word_count / total_expected_words if total_expected_words > 0 else 0  # Accuracy here is same as recall

    return accuracy, precision, recall

# Function to perform speech recognition on WAV files
def fn_speech_recognition_wav(wav_files, expected_phrases):
    try:
        r = sr.Recognizer()
        recognized_phrases = []

        # Iterate through the audio files
        for wav_file in wav_files:
            with sr.AudioFile(f"assets/Audios/{wav_file}") as source:
                audio = r.record(source)

            # Use the Sphinx recognizer
            try:
                recognized_phrase = r.recognize_sphinx(audio)
                recognized_phrases.append(recognized_phrase)
                print(f"Recognized: {recognized_phrase}")
            except sr.UnknownValueError:
                print("Sphinx could not understand the audio.")
                recognized_phrases.append("")
            except sr.RequestError as e:
                print(f"Sphinx error; {e}")
                recognized_phrases.append("")

        accuracy, precision, recall = calculate_word_level_metrics(expected_phrases, recognized_phrases)
        return accuracy, precision, recall

    except Exception as e:
        print(f"An error occurred: {e}")

# Define the test phrases (same for all environments)
expected_phrases = [
    "Artificial intelligence is the future of technology.",
    "Speech recognition systems are evolving rapidly.",
    "Can you understand what I am saying?",
    "Machine learning algorithms improve over time."
]

# Define the audio files for each environment
wav_files = {
    'Clean Speech': ['Clean-Noise-WAV-1.wav', 'Clean-Noise-WAV-2.wav', 'Clean-Noise-WAV-3.wav', 'Clean-Noise-WAV-4.wav'],
    'Moderate Noise': ['Moderate-Noise-WAV-1.wav', 'Moderate-Noise-WAV-2.wav', 'Moderate-Noise-WAV-3.wav', 'Moderate-Noise-WAV-4.wav'],
    'High Noise': ['High-Noise-WAV-1.wav', 'High-Noise-WAV-2.wav', 'High-Noise-WAV-3.wav', 'High-Noise-WAV-4.wav']
}

# Run speech recognition for each environment and display the results
print("Testing in Clean Speech Environment:")
accuracy, precision, recall = fn_speech_recognition_wav(wav_files['Clean Speech'], expected_phrases)
print(f"Word-level Accuracy: {accuracy*100:.2f}% | Precision: {precision*100:.2f}% | Recall: {recall:.2f}")

print("\nTesting in Moderate Noise Environment:")
accuracy, precision, recall = fn_speech_recognition_wav(wav_files['Moderate Noise'], expected_phrases)
print(f"Word-level Accuracy: {accuracy*100:.2f}% | Precision: {precision*100:.2f}% | Recall: {recall:.2f}")

print("\nTesting in High Noise Environment:")
accuracy, precision, recall = fn_speech_recognition_wav(wav_files['High Noise'], expected_phrases)
print(f"Word-level Accuracy: {accuracy*100:.2f}% | Precision: {precision*100:.2f}% | Recall: {recall:.2f}")
```

**10.2.1 Speech Recognition Using Computer Microphone**

Initially, an alternative versión for the speech recognition analyisis included using the computer’s built-in microphone with the SpeechRecognition library. However, it was discarded and replaced by the version with 'wav' files, although it is fully functional and is included as a complement to the research.. 

In this version, the system listens to the user’s spoken input, transcribes it using the Sphinx recognizer, and calculates the word-level accuracy, precision, and recall.

- **Microphone Input**: This implementation uses the system’s microphone to capture the user's speech. It identifies the specific microphone (e.g., "MacBook Pro (micrófono)") from the list of available devices.
  
- **Speech Recognition**: The captured audio is processed using the Sphinx recognizer, and each recognized phrase is compared with the expected phrase.

- **Metrics Calculation**: After obtaining the recognized phrases, the system calculates word-level accuracy, precision, and recall based on the comparison between the expected and recognized words. This is achieved through the `calculate_word_level_metrics` function, which compares words in the expected and recognized phrases using the `Counter` method.

```python
import speech_recognition as sr
from collections import Counter

# Function to calculate word-level metrics
def calculate_word_level_metrics(expected_phrases, recognized_phrases):
    correct_word_count = 0
    total_expected_words = 0
    total_recognized_words = 0

    for expected, recognized in zip(expected_phrases, recognized_phrases):
        expected_words = expected.lower().split()
        recognized_words = recognized.lower().split()

        # Count correct words
        expected_counter = Counter(expected_words)
        recognized_counter = Counter(recognized_words)
        
        common_words = expected_counter & recognized_counter
        correct_word_count += sum(common_words.values())

        # Total words in expected and recognized phrases
        total_expected_words += len(expected_words)
        total_recognized_words += len(recognized_words)

    # Calculate word-level precision, recall, and accuracy
    precision = correct_word_count / total_recognized_words if total_recognized_words > 0 else 0
    recall = correct_word_count / total_expected_words if total_expected_words > 0 else 0
    accuracy = correct_word_count / total_expected_words if total_expected_words > 0 else 0  # Accuracy here is same as recall

    return accuracy, precision, recall

# Function to perform speech recognition using microphone
def fn_speech_recognition_mic(expected_phrases):
    try:
        # List available microphones and select the desired one (e.g., MacBook Pro microphone)
        microphones = sr.Microphone.list_microphone_names()
        mic_index = None
        for index, name in enumerate(microphones):
            if name == "MacBook Pro (micrófono)":
                mic_index = index
                break
        
        if mic_index is None:
            print("Desired microphone not found. Exiting...")
            return

        # Creating a recognition object
        r = sr.Recognizer()
        recognized_phrases = []

        # Loop through the expected phrases and capture spoken input via microphone
        for expected in expected_phrases:
            with sr.Microphone(device_index=mic_index) as source:
                print(f"Please say: '{expected}'")
                audio = r.listen(source, timeout=0, phrase_time_limit=5)

            # Recognize speech using Sphinx (offline)
            try:
                recognized_phrase = r.recognize_sphinx(audio)
                recognized_phrases.append(recognized_phrase)
                print(f"Recognized: {recognized_phrase}")
            except sr.UnknownValueError:
                print("Sphinx could not understand the audio.")
                recognized_phrases.append("")
            except sr.RequestError as e:
                print(f"Sphinx error; {e}")
                recognized_phrases.append("")

        # Calculate metrics (accuracy, precision, recall)
        accuracy, precision, recall = calculate_word_level_metrics(expected_phrases, recognized_phrases)
        print(f"Word-level Accuracy: {accuracy*100:.2f}% | Precision: {precision:.2f} | Recall: {recall:.2f}")

    except sr.WaitTimeoutError:
        print("Listening timed out while waiting for phrase to start.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Define the test phrases (same phrases used in other environments)
expected_phrases = [
    "Artificial intelligence is the future of technology.",
    "Speech recognition systems are evolving rapidly.",
    "Can you understand what I am saying?",
    "Machine learning algorithms improve over time."
]

# Run the speech recognition using the microphone
fn_speech_recognition_mic(expected_phrases)
```
